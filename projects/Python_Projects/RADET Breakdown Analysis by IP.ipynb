```python
import pandas as pd
import re
import os
```

```python
# Define the folder path (RADET folder on the desktop)
folder_path = os.path.expanduser('/Users/seyiabolarin/Desktop/Python Workbook')

# Ensure the folder exists
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
# Load the first RADET Excel file
radet1 = pd.read_excel('/Users/seyiabolarin/Desktop/Uncle Jonathan/ACE1_ACE3_ACE5_TMEC RISE_Combined RADET 26.07.24.xlsx')
```

```python
# Load the second RADET Excel file
radet2 = pd.read_excel('/Users/seyiabolarin/Desktop/Uncle Jonathan/ACE 2_ACE 4_ACE-6_KPCARE 1_KPCARE 2_RADET_26-07-2024.xlsx')
```

```python
# Concatenate the DataFrames
merged_radet = pd.concat([radet1, radet2], ignore_index=True)
```

```python
radet_list = list(merged_radet.columns)
print(radet_list)
```

```python
# Select only the 'state' and 'date of birth' columns
radet_selected = merged_radet[['IP','State', 'L.G.A','Facility Name','Patient ID','Hospital Number','NDR Patient Identifier','ART Start Date (yyyy-mm-dd)','Last Pickup Date (yyyy-mm-dd)','Months of ARV Refill','Regimen at ART Start','Current ART Regimen','Date of Viral Load Sample Collection (yyyy-mm-dd)','Date of Current ViralLoad Result Sample (yyyy-mm-dd)','Current Viral Load (c/ml)','Date of Current Viral Load (yyyy-mm-dd)','Current ART Status','Date of Current ART Status','Client Verification Outcome']]
```

```python
radet_selected
```

```python
#converting all date to date format
radet_selected['ART Start Date (yyyy-mm-dd)'] = pd.to_datetime(radet_selected['ART Start Date (yyyy-mm-dd)'], errors='coerce')
radet_selected['Last Pickup Date (yyyy-mm-dd)'] = pd.to_datetime(radet_selected['Last Pickup Date (yyyy-mm-dd)'], errors='coerce')
radet_selected['Date of Viral Load Sample Collection (yyyy-mm-dd)'] = pd.to_datetime(radet_selected['Date of Viral Load Sample Collection (yyyy-mm-dd)'], errors='coerce')
radet_selected['Date of Current ViralLoad Result Sample (yyyy-mm-dd)'] = pd.to_datetime(radet_selected['Date of Current ViralLoad Result Sample (yyyy-mm-dd)'], errors='coerce')
radet_selected['Date of Current Viral Load (yyyy-mm-dd)'] = pd.to_datetime(radet_selected['Date of Current Viral Load (yyyy-mm-dd)'], errors='coerce')
radet_selected['Date of Current ART Status'] = pd.to_datetime(radet_selected['Date of Current ART Status'], errors='coerce')
```

```python
# Step 2: Combine the folder path with the filename
file_path = os.path.join(folder_path, 'radet_selected.csv')
```

```python
# Step 3: Save the file to the specified folder
radet_selected.to_csv(file_path, index=False)
```

```python
# Get the unique values in the 'IP' column
ip_values = radet_selected['IP'].unique()

# Loop through each unique value in the 'IP' column and save to a separate file
for ip in ip_values:
    # Filter the DataFrame for the current IP value
    radet_filtered = radet_selected[radet_selected['IP'] == ip]
    
    # Create a file name based on the IP value
    file_name = f'{ip}.xlsx'
    
    # Combine the folder path with the file name
    file_path = os.path.join(folder_path, file_name)
    
    # Save the filtered DataFrame to an Excel file in the RADET folder
    radet_filtered.to_excel(file_path, index=False)
    
    # Print a message indicating the file has been saved
    print(f'Saved file: {file_path}')
```
